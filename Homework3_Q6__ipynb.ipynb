 In a multiple regression model: ùê≤ = ùêóùõÉ + ùõú, it is critical to know if (ùêóùëáùêó)‚àí1 exists. The diagonal elements of (ùêóùëáùêó)‚àí1 in correlation form, i.e., ùêó is normalized, are often called Variance Inflation Factors (VIFs), and they are important multicollinearity diagnostic. VIF for the ùëóùë°h regression coefficient is expressed as
VIFùëó= 1/1‚àíùëÖ2 , where ùëÖ2 is the coefficient of multiple determination obtained from regressing ùê± on the other regressor ùëóùëó
variables (ùê±1 to ùê±ùëù, except ùê±ùëó). Calculate all the VIFs in the autompg dataset and discuss your observation.

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor 
import statsmodels.api as sm
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/homework_3/auto_data.csv', sep=";", header=None, names=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',  'acceleration', 'model year', 'origin', 'car name' ])

df.head()


```

```python
X = df[['mpg','cylinders', 'displacement', 'horsepower', 'weight',  'acceleration', 'model year', 'origin']]
X
```

```python
X = sm.add_constant(X)
```

```python
VIF = pd.DataFrame()
VIF["Data Type - Var"] = X.columns
```

```python
VIF
```

```python
VIF["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
```
